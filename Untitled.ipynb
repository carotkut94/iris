```python
# Basic step for using scikit-learn library
# 1. Choose a class of model
# 2. Choose model hyperparameter
# 3. Arrange data into features matrix and target array
# 4. Fit model to data
# 5. Apply trained model to new data
```

```python
# Lets choose simple linear regression
x = 10 * np.random.rand(100)
```

```python
y = 3 * x + np.random.rand(100)
```

```python
plt.scatter(x,y) # Scatter plot for x and y
```

```python
from sklearn.linear_model import LinearRegression # Importing linear regression
```

```python
model = LinearRegression(fit_intercept=True)
```

```python
model
```

```python
X = x.reshape(-1,1)
X.shape
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
x_fit = np.linspace(-1,11)
```

```python
X_fit = x_fit.reshape(-1,1)
```

```python
yfit = model.predict(X_fit)
```

```python
plt.scatter(x,y)
plt.plot(x_fit, yfit)
```
